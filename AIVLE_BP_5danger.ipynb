#### preprocessing

```python
os.mkdir('/content/Dataset/train/Fall(A)')
os.mkdir('/content/Dataset/train/Drop(B)')
os.mkdir('/content/Dataset/train/Crush(C)')
os.mkdir('/content/Dataset/train/Fire(D)')
os.mkdir('/content/Dataset/train/Electric(E)')

os.mkdir('/content/Dataset/val/Fall(A)')
os.mkdir('/content/Dataset/val/Drop(B)')
os.mkdir('/content/Dataset/val/Crush(C)')
os.mkdir('/content/Dataset/val/Fire(D)')
os.mkdir('/content/Dataset/val/Electric(E)')
```

```python
PATH_data = '/content/drive/MyDrive/AIVLE/AIHUB_danger/Validation/X'
DATA_main = os.listdir(PATH_data)

danger = []
for i in range(len(DATA_main)):
    if DATA_main[i].endswith('.jpg'): danger.append(DATA_main[i])
len(danger)
```

```python
from sklearn.model_selection import train_test_split
train, val = train_test_split(danger, test_size = 0.2, random_state = 42)
len(train), len(val)
```

```python
import shutil
```

```python
for i in range(len(train)):
    if (train[i].split('_'))[2][0] == 'N':
        if (train[i].split('_'))[1][0] == 'A':
            shutil.copy(PATH_data + '/' + train[i], '/content/Dataset/train/Fall(A)')
        elif (train[i].split('_'))[1][0] == 'B':
            shutil.copy(PATH_data + '/' + train[i], '/content/Dataset/train/Drop(B)')
        elif (train[i].split('_'))[1][0] == 'C':
            shutil.copy(PATH_data + '/' + train[i], '/content/Dataset/train/Crush(C)')
        elif (train[i].split('_'))[1][0] == 'D':
            shutil.copy(PATH_data + '/' + train[i], '/content/Dataset/train/Fire(D)')
        else:
            shutil.copy(PATH_data + '/' + train[i], '/content/Dataset/train/Electric(E)')

for i in range(len(val)):
    if (val[i].split('_'))[2][0] == 'N':
        if (val[i].split('_'))[1][0] == 'A':
            shutil.copy(PATH_data + '/' + val[i], '/content/Dataset/val/Fall(A)')
        elif (val[i].split('_'))[1][0] == 'B':
            shutil.copy(PATH_data + '/' + val[i], '/content/Dataset/val/Drop(B)')
        elif (val[i].split('_'))[1][0] == 'C':
            shutil.copy(PATH_data + '/' + val[i], '/content/Dataset/val/Crush(C)')
        elif (val[i].split('_'))[1][0] == 'D':
            shutil.copy(PATH_data + '/' + val[i], '/content/Dataset/val/Fire(D)')
        else:
            shutil.copy(PATH_data + '/' + val[i], '/content/Dataset/val/Electric(E)')
```
